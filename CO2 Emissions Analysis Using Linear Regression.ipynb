### CO2 Emission Results From Several Categories

#### Vehicle Class Category

```python
df2.columns
```

```python
df2 = df2.rename(columns={'Vehicle Class':'vehicleClass', 'Engine Size(L)': 'engineSize', 'Cylinders':'cylinders',
                         'Transmission':'transmission', 'Fuel Type':'fuelType', 'Fuel Consumption Comb (L/100 km)':'fuelConsumptionCombination(L/100km)'})
df2
```

```python
import matplotlib.pyplot as plt

# Assuming df_final is your DataFrame
top10 = df2.groupby('vehicleClass')['CO2 Emissions(g/km)'].mean().nlargest(10)

# Plotting the bar chart
ax = top10.plot(kind='bar', figsize=(10, 8))
plt.title('Top 10 Vehicle Classes by Mean CO2 Emissions')
plt.xlabel('Vehicle Class')
plt.ylabel('Mean CO2 Emissions(g/km)')

for i, v in enumerate(top10):
  ax.text(i, v + 0.1 , f'{v:.2f}', ha='center', va='bottom')
plt.show()
```

#### Engine Size and Cylinders Category

```python
import matplotlib.pyplot as plt

# Assuming df_final is your DataFrame
top10 = df2.groupby(['engineSize', 'cylinders'])['CO2 Emissions(g/km)'].mean().nlargest(10)

# Plotting the bar chart
ax = top10.plot(kind='bar', figsize=(10, 8))
plt.title('Top 10 Engine Size and Cylinders Impact by Mean CO2 Emissions')
plt.xlabel('Engine Size and Cylinders')
plt.ylabel('Mean CO2 Emissions(g/km)')

for i, v in enumerate(top10):
  ax.text(i, v + 0.1 , f'{v:.2f}', ha='center', va='bottom')
plt.show()
```

### Modeling

#### Model 1

Using dataframe2 (df2)

Focused on Engine Size, Cylinders, and Fuel Consumption Combination to see how they impact to CO2 Emission without One Hot Encoding.

```python
df2.columns
```

```python
x = df2[['engineSize', 'cylinders', 'fuelConsumptionCombination(L/100km)']]
x
```

```python
y = df2['CO2 Emissions(g/km)']
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(x_train, y_train)

y_pred = model1.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate RMSE score
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"Root Mean Squared Error (RMSE): {rmse}")

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2}")
```

### model 2

Using dataframe1 (df)

```python
df.columns
```

```python
x = df[['Engine Size(L)', 'Cylinders', 'Fuel Consumption City (L/100 km)',
       'Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)',
       'Fuel Consumption Comb (mpg)']]
x
```

```python
y = df['CO2 Emissions(g/km)']
y
```

```python
from sklearn.model_selection import train_test_split

# Split data for train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model2 = LinearRegression()
model2.fit(x_train, y_train)

# Predict y
y_pred = model2.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate RMSE score
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"Root Mean Squared Error (RMSE): {rmse}")

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2): {r2}")
```

In this test, it is found that model2 tends to be lower for RMSE, MSE, MAE, and R-squared values compared to model1.
